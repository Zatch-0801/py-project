```python
import numpy as np
import matplotlib.pyplot as plt

#sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)
#generate output y using function
y = sigmoid(x)

print("First 5 sigmoid values:")
print(y[:5])

#plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 400)
# tanh function
y = tanh(x)
print("First 5 tanh values:")
print(y[:5])

# plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Tanh Function', color='green')
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#ReLU function
def relu(x):
    return np.maximum(0, x)

x = np.linspace(0, 40)
# Compute y values using the ReLU function
y = relu(x)
print("First 5 ReLU values:")
print(y[:5])
# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='ReLU Function', color='red')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Generate x values from 0 to 10
x = np.linspace(0, 10, 400)
# Compute y values using the Leaky ReLU function
y = leaky_relu(x)

# Print the first 5 Leaky ReLU values
print("First 5 Leaky ReLU values:")
print(y[:5])

#plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Leaky ReLU Function', color='purple')
plt.title('Leaky ReLU Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
